```python
import numpy as np
from math import factorial
import matplotlib.pyplot as plt
```

```python
def random_walk(N, p):
        N = int(N)
        distribution = np.empty(N + 1)
    
        for n in range(N + 1):
            binomial = factorial(N) / (factorial(n) * factorial(N - n))
            distribution[n] = binomial * (p ** n) * (1 - p) ** (N - n)
        return distribution
```

```python
def expectation_value(P):
    ev = 0
    for idx, num in enumerate(P):
        ev += (num + 1) * num
    return ev
```

```python
def binomial_sampling(N,p,samples):
    counts, bins = np.histogram(np.random.binomial(N,p, samples), bins=np.arange(0,N), density=True)
    return counts, bins
```

```python
dist1=random_walk(10,0.5)
ev_1=expectation_value(dist1)
dist2=random_walk(10,0.1)
ev_2=expectation_value(dist2)
dist3=random_walk(60,0.5)
ev_3=expectation_value(dist3)
dist4=random_walk(60,0.8)
ev_4=expectation_value(dist4)
fig1, ax = plt.subplots(2, 1)

ax[0].plot(dist1, label="(N,p):(10,0.5), "+"<E>="+str(np.floor(ev_1)))
ax[0].plot(dist2, label="(N,p):(10,0.1), "+"<E>="+str(np.floor(ev_2)))
ax[0].legend()
ax[0].set_xlabel("Steps [-]")
ax[0].set_ylabel("Probability [-]")

ax[1].plot(dist3, label="(N,p):(60,0.5), "+"<E>="+str(np.floor(ev_3)))
ax[1].plot(dist4, label="(N,p):(60,0.8), "+"<E>="+str(np.floor(ev_4)))
ax[1].legend()
ax[1].set_xlabel("Steps [-]")
ax[1].set_ylabel("Probability [-]")

fig1.tight_layout()
fig1.savefig('Week1_Student.png',bbox_inches='tight', format='png',dpi=600)
```

```python
your_probabilities=random_walk(200,0.5)
counts,bins = binomial_sampling(200, 0.5, 10000)
fig=plt.figure()
plt.stairs(counts,bins,label='Random Generator')
plt.plot(your_probabilities,label='Your Function')
plt.xlabel('Steps to the right [-]')
plt.ylabel('Probability [-]')
plt.legend()
fig.tight_layout()
```
