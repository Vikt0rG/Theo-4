In this exercise, we focus on the implementation of Fermi-Dirac, Bose-Einstein and Maxwell-Boltzmann distribution functions.
<br>


> ### **Introduction: Fermi-Dirac Distribution**

<br>


The Fermi-Dirac distribution characterizes the statistical distribution of fermions and is defined as: 

$$ f_{\text{FD}}(E) = \frac{1}{\exp\left(\frac{E - \mu}{k T}\right) + 1} \tag{Equation 1}$$ 

Where:
- $ f_{\text{FD}}(E) $: Fermi-Dirac distribution function.
- $ E $: Energy of the system.
- $ \mu $: Chemical potential.
- $ k $: Boltzmann constant.
- $ T $: Temperature.



<br>


> ### **Introduction: Bose-Einstein Distribution**

<br>


The Bose-Einstein distribution characterizes the statistical distribution of bosons and is defined as: 

$$ f_{\text{BE}}(E) = \frac{1}{\exp\left(\frac{E - \mu}{k T}\right) - 1} \tag{Equation 2}$$ 

Where:
- $ f_{\text{BE}}(E) $: Bose-Einstein distribution function.
- $ E $: Energy of the system.
- $ \mu $: Chemical potential.
- $ k $: Boltzmann constant.
- $ T $: Temperature.

These distributions are essential for understanding the behavior of fermionic and bosonic particles in a variety of physical systems.

<br>


> ### **Introduction: Maxwell-Boltzmann Distribution**

<br>


The Maxwell-Boltzmann distribution characterizes the statistical distribution of classical particles and is defined as: 

$$ f_{\text{MB}}(E) = \exp\left(\frac{-(E - \mu)}{k T}\right)  \tag{Equation 3}$$ 

Where:
- $ f_{\text{MB}}(E) $: Maxwell-Distribution distribution function.
- $ E $: Energy of the system.
- $ \mu $: Chemical potential.
- $ k $: Boltzmann constant.
- $ T $: Temperature.


<br>

> ### **Introduction: Classical Limit Convergence**

<br>

$$ (E - \mu) \gt\gt k T \tag{Equation 4}$$ 

The distribution functions in equations 1 to 3 are identical in the classical limiting case, which is stated in equation 4.


<br>

> ### **Introduction: Particle Number Calculation**

<br>

To calculate the number of particles in the system, we integrate the distribution functions multiplied by the density of states $ g(E)$. 

$$ N = \int_{\text{Lower limit}}^{\text{Upper limit}} g(E) \cdot f(E) \, dE \tag{Equation 5} $$




In this case, we will make use of both free fermionic and bosonic systems, where we can use a density of states proportional to the square root of energy. The integral will have a form that is given by (neglecting normalization):

$$ N = \int_{\text{Lower limit}}^{\text{Upper limit}} \sqrt{E} \cdot f(E) \, dE \tag{Equation 6} $$

Where $ f(E)$ are the distributions in equations 1 and 2.

Adjust the integration limits based on the specific physical system under consideration.

2. **Implement the `maxwell_boltzmann_distribution` function**:
    - The function takes three parameters: `energy`, `temperature`, and `mu` (chemical potential).
    - It returns the Maxwell Boltzmann distribution for the given parameters, calculated according to Equation 3.

> Note: Ensure that you use the Boltzmann constant in eV/K.




```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Boltzmann constant in eV
k_boltzmann_eV = 8.617333262145e-5

def fermi_dirac_distribution(energy, temperature, mu):
    return 1 / (np.exp((energy - mu) / (k_boltzmann_eV * temperature)) + 1)

def bose_einstein_distribution(energy, temperature, mu):
    return 1 / (np.exp((energy - mu) / (k_boltzmann_eV * temperature)) - 1)

def maxwell_bolztmann(energy, temperature, mu):
    return np.exp(-(energy - mu) / (k_boltzmann_eV * temperature))
```

In this task, you will implement two functions: `integrand` and `integrate_square_root_energy` and generate one plot.

1. **Define the  `integrand` function**:
    - Takes `energy`, `temperature`, `mu`, and `distribution_function` as parameters.
    - Returns the integrand in equation 6.
      
2. **Define the `integrate_square_root_energy` function**:
    - Takes `temperature`, `mu`, `distribution_function`, `lower_limit`, and `upper_limit` as parameters.
    - Uses `quad` from `scipy`, and the  `integrand` function to perform the integration.
    - Returns the result of the integration in equation 6.

3. **Plot results**:
    - Plot the particle number for both systems as a function of temperature from 0.1 to 30000K
    - As always label your plots correctly.

>Note: Use the same energy ranges for the integrals as the ones used to generate the 3D plots. 

```python
from scipy.integrate import quad

def integrand(energy, temperature, mu, distribution_function):
    return np.sqrt(energy) * distribution_function(energy, temperature, mu)

def integrate_square_root_energy(temperature, mu, distribution_function, lower_limit, upper_limit):
    return quad(lambda u: integrand(u, temperature, mu, distribution_function), lower_limit, upper_limit)[0]

N_fermi = np.vectorize(integrate_square_root_energy)(t, 5, fermi_dirac_distribution, 0, 10)
N_bose = np.vectorize(integrate_square_root_energy)(t, -1, bose_einstein_distribution, 0, 10)

plt.plot(t, N_fermi, label='Fermi-Dirac')
plt.plot(t, N_bose, label='Bose-Einstein')

plt.xlabel('Temperature (K)')
plt.ylabel('Particle Number')
plt.title('Particle Number vs Temperature')
plt.legend()
```
